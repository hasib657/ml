# Exercise 1: Array Creation and Modification (NumPy)

```python
# Exercise 1: Array Creation and Modification (NumPy)# Exercise 1: 4x4 matrix of random ints (10â€“50), set >40 to 0
import numpy as np

# Create 4x4 array with random integers between 10 and 50 (inclusive)
array = np.random.randint(10, 51, size=(4, 4))  # 51 because upper bound is exclusive
print("Original array:\n", array)

# Modify: set all values > 40 to 0
array[array > 40] = 0
print("\nModified array (values > 40 set to 0):\n", array)
```

# Exercise 2: Statistical Operations (NumPy)

```python
# Solution 2: Statistical Operations
# Task: Modify the code to compute the median and standard deviation instead of the mean, and
# apply it to a 1D array of 20 random floats between 0 and 10
array = np.random.uniform(0, 10, size=20)
median_value = np.median(array)
std_deviation = np.std(array)
print("Array:", array)
print("Median: ", median_value)
print("Standard Deviation: ", std_deviation)
```

# Exercise 3: DataFrame Filtering (Pandas)

```python
# Exercise 3: Filter age > 25 and city == "New York"
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 20],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago']
}
df = pd.DataFrame(data)

# Filter: age > 25 AND city == "New York"
filtered_df = df[(df['age'] > 25) & (df['city'] == 'New York')]

print("Filtered DataFrame:\n", filtered_df)
```

# Exercise 4: Grouping and Aggregation (Pandas)

```python
# Exercise 4: Group by city, compute max age
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 20, 28],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']
}
df = pd.DataFrame(data)

# Group by city and get max age
max_age_by_city = df.groupby('city')['age'].max()

print("Maximum age by city:\n", max_age_by_city)
```

# Exercise 5: Boxplot Customization (Seaborn)

```python
# Exercise 5: Boxplot of total_bill by day, colored by sex + title
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Boxplot: total_bill by day, colored by sex
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')

# Add title
plt.title("Total Bill by Day and Sex")
plt.show()
```

# Exercise 6: Correlation Heatmap (Seaborn)

```python
# Exercise 6: Heatmap of correlation matrix (numerical cols only)
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Select numerical columns only
numerical_tips = tips.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numerical_tips.corr()

# Heatmap with annotations and coolwarm colormap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Exercise 7: Interactive Scatter Plot (Plotly)

```python
# Exercise 7: Scatter total_bill vs tip, color=day, size=size, hover=time+smoker
import plotly.express as px

tips = px.data.tips()

fig = px.scatter(
    tips,
    x='total_bill',
    y='tip',
    color='day',           # Color by day
    size='size',           # Size by table size
    hover_data=['time', 'smoker'],  # Extra hover info
    title="Total Bill vs Tip (Interactive)"
)
fig.show()
```

# Exercise 8: Bar Plot with Customization (Plotly)

```python
# Exercise 8: Avg total_bill per day, bars colored by sex
import plotly.express as px
import pandas as pd

tips = px.data.tips()

# Compute average total_bill per day and sex
agg_data = tips.groupby(['day', 'sex'])['total_bill'].mean().reset_index()

fig = px.bar(
    agg_data,
    x='day',
    y='total_bill',
    color='sex',
    barmode='group',  # Side-by-side bars
    title="Average Total Bill by Day and Sex"
)
fig.show()
```
